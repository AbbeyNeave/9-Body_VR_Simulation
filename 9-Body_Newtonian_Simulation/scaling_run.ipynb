## Set Up

```python
# import libraries & modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import body_class
import create_vars_fcns
```

```python
# read in data to be scaled
data = pd.read_csv('input_data_notscaled.csv')
```

```python
# define all 9 bodies as members of body class
sun = body_class.Body(mass = data.iloc[0,1], initial_pos = [data.iloc[0,2],data.iloc[0,3],data.iloc[0,4]], initial_vel = [data.iloc[0,5],data.iloc[0,6],data.iloc[0,7]])
merc = body_class.Body(mass = data.iloc[1,1], initial_pos = [data.iloc[1,2],data.iloc[1,3],data.iloc[1,4]], initial_vel = [data.iloc[1,5],data.iloc[1,6],data.iloc[1,7]])
ven = body_class.Body(mass = data.iloc[2,1], initial_pos = [data.iloc[2,2],data.iloc[2,3],data.iloc[2,4]], initial_vel = [data.iloc[2,5],data.iloc[2,6],data.iloc[2,7]])
earth = body_class.Body(mass = data.iloc[3,1], initial_pos = [data.iloc[3,2],data.iloc[3,3],data.iloc[3,4]], initial_vel = [data.iloc[3,5],data.iloc[3,6],data.iloc[3,7]])
mars = body_class.Body(mass = data.iloc[4,1], initial_pos = [data.iloc[4,2],data.iloc[4,3],data.iloc[4,4]], initial_vel = [data.iloc[4,5],data.iloc[4,6],data.iloc[4,7]])
jupit = body_class.Body(mass = data.iloc[5,1], initial_pos = [data.iloc[5,2],data.iloc[5,3],data.iloc[5,4]], initial_vel = [data.iloc[5,5],data.iloc[5,6],data.iloc[5,7]])
saturn = body_class.Body(mass = data.iloc[6,1], initial_pos = [data.iloc[6,2],data.iloc[6,3],data.iloc[6,4]], initial_vel = [data.iloc[6,5],data.iloc[6,6],data.iloc[6,7]])
uranus = body_class.Body(mass = data.iloc[7,1], initial_pos = [data.iloc[7,2],data.iloc[7,3],data.iloc[7,4]], initial_vel = [data.iloc[7,5],data.iloc[7,6],data.iloc[7,7]])
nept = body_class.Body(mass = data.iloc[8,1], initial_pos = [data.iloc[8,2],data.iloc[8,3],data.iloc[8,4]], initial_vel = [data.iloc[8,5],data.iloc[8,6],data.iloc[8,7]])
```

```python
# define bodies as objects list
bodies_as_objects_list = [sun, merc, ven, earth, mars, jupit, saturn, uranus, nept]
bodies_as_objects_list_without_sun = [merc, ven, earth, mars, jupit, saturn, uranus, nept]
```

## Creating Scaling Factor Array

### Visualising Real Body Distances from Sun

```python
# get between body position vector array
between_body_position_vector_array = create_vars_fcns.create_3d_between_body_position_vector_array(bodies_as_objects_list) #shape (9, 9, 3, 18262)

# select position vectors from sun to other bodies only
position_vectors_from_sun_with_sun_array = between_body_position_vector_array[0, :, :, 0]  #shape (9, 3)

# get each body's distance from sun in km
distances_from_sun_km_with_sun_array = np.empty(9)
for body_index in np.array((range(9))):
    distances_from_sun_km_with_sun_array[body_index] = np.linalg.norm([position_vectors_from_sun_with_sun_array[body_index,0], position_vectors_from_sun_with_sun_array[body_index,1],position_vectors_from_sun_with_sun_array[body_index,2]])

# convert body's distance from sun to meters
distances_from_sun_meters_with_sun_array = distances_from_sun_km_with_sun_array * 1000
```

```python
# graph planet's real distances from the sun
labels = ["sun","merc", "ven", "earth", "mars", "jupit", 'sat', 'uran', 'nept',]
x = np.arange(len(labels))
fig, ax = plt.subplots()
planet = ax.bar(x, distances_from_sun_meters_with_sun_array)
ax.set_title("Real Distances From Sun")
ax.set_ylabel("Real Distance (m)")
ax.set_xticks(x)
ax.set_xticklabels(labels)
leg = ax.legend()
```

### Find Scaling Factors by Visualising Body Distances from Sun with Trial & Error

```python
# define scaling factors array (made by visualising with trial and error)
scaling_factors_with_sun = np.array([1, 1.080e+01, 9.900e+00, 8.100e+00, 6.300e+00, 3.222e+00, 2.007e+00, 1.260e+00, 9.900e-01])
scaling_factors_without_sun = np.array([1.080e+01, 9.900e+00, 8.100e+00, 6.300e+00, 3.222e+00, 2.007e+00, 1.260e+00, 9.900e-01])
```

```python
# scale planet distances from Sun
scaled_distances_from_sun_meters_with_sun_array = distances_from_sun_meters_with_sun_array * scaling_factors_with_sun
```

```python
# graph planet's real distances from the sun
labels = ["sun","merc", "ven", "earth", "mars", "jupit", 'sat', 'uran', 'nept',]
x = np.arange(len(labels))  # the label locations
fig, ax = plt.subplots()
planet = ax.bar(x, scaled_distances_from_sun_meters_with_sun_array)
ax.set_title("Distances as Scaled for VR")
ax.set_ylabel("Scaled Distance (m)")
ax.set_xticks(x)
ax.set_xticklabels(labels)
leg = ax.legend()
```

## Scaling Initial Positions (*potential debug*)

```python
# get between body unit position vector array 
between_body_unit_position_vector_array = create_vars_fcns.create_3d_between_body_unit_vector_array(between_body_position_vector_array) #shape (9, 9, 3, 18262)
### POTENTIAL DEBUG: BELOW OUTPUT SOMETIMES SWITCHES TO 0's/NANs, SEEMINGLY RANDOMLY 
# print(f'Check for bug - between_body_unit_position_vector_array[0,:,:,0] = {unit_position_vectors_from_sun_with_sun_array}')

# select unit position vectors from sun to other bodies only
unit_position_vectors_from_sun_with_sun_array = between_body_unit_position_vector_array[0,:,:,0] #shape (9, 3)
```

```python
# create and fill array of scaled initial position vectors from sun to all other bodies
scaled_initial_position_vectors_from_sun_meters_with_sun = np.empty((9,3))
for body_index in np.array((range(9))):
    scaled_initial_position_vectors_from_sun_meters_with_sun[body_index,:] = unit_position_vectors_from_sun_with_sun_array[body_index] * scaled_distances_from_sun_meters_with_sun_array[body_index]
```

## Scaling Initial Velocities

```python
# find magnitudes of real initial velocity vectors
initial_vecloity_magnitudes = []
for body in bodies_as_objects_list_without_sun:
    initial_vecloity_magnitudes.append(np.linalg.norm(body.vel_3d_array[:, 0]))
initial_vecloity_magnitudes = np.array(initial_vecloity_magnitudes) # of shape (8)
```

```python
# find unit vectors for real initial velocity vectors
initial_vecloity_unit_vectors = []
for body in bodies_as_objects_list_without_sun:
    initial_vecloity_unit_vectors.append(body.vel_3d_array[:, 0]/np.linalg.norm(body.vel_3d_array[:, 0]))
initial_vecloity_unit_vectors = np.array(initial_vecloity_unit_vectors)  # of shape (8, 3)
```

```python
# scale magnitudes of real initial velocity vectors using Kepler's Law (* np.sqrt(1/scaling_factor))
initial_vecloity_magnitudes_scaled = initial_vecloity_magnitudes * np.sqrt(1/scaling_factors_without_sun) # of shape (8)
```

```python
# find scaled initial velocity vectors
initial_velocity_vectors_scaled = np.empty((8, 3))
for body_index in np.array(range(8)):
    initial_velocity_vectors_scaled[body_index, :] = initial_vecloity_unit_vectors[body_index, :] * initial_vecloity_magnitudes_scaled[body_index]
```

```python
# include sun's initial velocity at [0,0,0] to initial_velocity_vectors_scaled
sun_initial_velocity = [0,0,0]
scaled_initial_velocity_vectors_with_sun = np.vstack([sun_initial_velocity, initial_velocity_vectors_scaled])
```

## Create CSV for Scaled Positions/Velocities Data

```python
#function to produce position vector
input_data_scaled = pd.DataFrame({
        "Body Name" : data.iloc[:,0],
        "Mass (kg)" : data.iloc[:,1],
        "x_position (m)": scaled_initial_position_vectors_from_sun_meters_with_sun[:,0],
        "y_position (m)": scaled_initial_position_vectors_from_sun_meters_with_sun[:,1],
        "z_position (m)": scaled_initial_position_vectors_from_sun_meters_with_sun[:,2],
        "x_velocity (m/s)": scaled_initial_velocity_vectors_with_sun[:,0],
        "y_velocity (m/s)": scaled_initial_velocity_vectors_with_sun[:,1],
        "z_velocity (m/s)": scaled_initial_velocity_vectors_with_sun[:,2]
        })
input_data_scaled.to_csv('input_data_scaled.csv', encoding='utf-8', index=False)
```

## Scaling Body Volumes

```python
# get real body diameters
real_body_diameters_meters = np.array(data)[:,8]
```

```python
# graph real body diameters
labels = ["sun","merc", "ven", "earth", "mars", "jupit", 'sat', 'uran', 'nept',]
x = np.arange(len(labels)) # the label locations
fig, ax = plt.subplots()
planet = ax.bar(x, real_body_diameters_meters)
ax.set_ylabel("Real Diameter (meters)")
ax.set_title("Real Body Diameters")
ax.set_xticks(x)
ax.set_xticklabels(labels)
leg = ax.legend()
```

```python
# define scaling factors array (made by visualising with trial and error)
scaling_factors_for_volumes = pd.Series([1,50,30,35,40,6.5,6.5,11.5,11])
```

```python
# scale body diameters
scaled_body_diameters_meters = real_body_diameters_meters * scaling_factors_for_volumes
```

```python
# graph scaled body diameters
labels = ["sun","merc", "ven", "earth", "mars", "jupit", 'sat', 'uran', 'nept',]
x = np.arange(len(labels))  # the label locations
fig, ax = plt.subplots()
planet = ax.bar(x, scaled_body_diameters_meters)
ax.set_ylabel("Scaled Diameter (meters)")
ax.set_title("Scaled Body Diameters")
ax.set_xticks(x)
ax.set_xticklabels(labels)
leg = ax.legend()
```

```python
# convert to scaled radii
scaled_body_radii_meters = scaled_body_diameters_meters/2

# scale for unity visuals 
scaled_body_radii_meters = scaled_body_radii_meters / 100000000
```

```python
#function to produce position vector
output_radii = pd.DataFrame({
        "Body Name" :  data.iloc[:,0],
        "Radii (m/100000000)" : scaled_body_radii_meters,
        })
output_radii.to_csv('output_radii.csv', encoding='utf-8', index=False)
```
